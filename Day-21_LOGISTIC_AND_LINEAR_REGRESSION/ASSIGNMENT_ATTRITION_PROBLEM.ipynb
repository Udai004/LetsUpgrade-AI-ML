## Dropping irrelevant attributes
Employee attrition is not effected by attributes such as as EmployeeCount and EmployeeID,moreover attributes with constant value for all the records such as Over18 and StandardHours does'nt affect the attrition rate.

```python

data1=data.drop(['EmployeeCount','EmployeeID','Over18','StandardHours'],axis=1)
data1.head()
```

```python
data1.isna().sum()
```

```python
data2=data1.fillna(data1.mean())
data2.isna().sum()
```

```python

from sklearn.preprocessing import LabelEncoder
encoder=LabelEncoder()
```

```python

data2['Attrition']=encoder.fit_transform(data1['Attrition'])
```

```python

data2['BusinessTravel']=encoder.fit_transform(data1['BusinessTravel'])
```

```python
data2['Department']=encoder.fit_transform(data1['Department'])
```

```python

data2['EducationField']=encoder.fit_transform(data1['EducationField'])
```

```python
data2['Gender']=encoder.fit_transform(data1['Gender'])
```

```python

data2['JobRole']=encoder.fit_transform(data1['JobRole'])
```

```python
data2['MaritalStatus']=encoder.fit_transform(data1['MaritalStatus'])
```

```python

data2.head()
```

## splitting the dataset

```python

y=data2.iloc[:,1:2]
y.head()
```

```python
x=data2.drop(['Attrition'],axis=1)
x.head()
```

# Logistic Regression

```python
import statsmodels.api as sm
```

```python
x1=sm.add_constant(x)
logistic=sm.Logit(y,x1)
result=logistic.fit()
result.summary()
```

2) Out of all these variables, namely Age, MaritalStatus, NumCompaniesWorked, TotalWorkingYears, TrainingTimesLastYear,YearsSinceLastPromotion and YearsWithCurrManager are most significant.
